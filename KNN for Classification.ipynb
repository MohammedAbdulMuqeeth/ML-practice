```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer = pd.read_csv('d:/dataset/cancerdata.csv')
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
cancer.drop(['id','Unnamed: 32'],axis=1, inplace=True)
```

```python
cancer.head()
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
Scaler.fit(cancer.drop('diagnosis',axis=1))
```

```python
Scaled_features = Scaler.transform(cancer.drop('diagnosis',axis=1))
```

```python
Scaled_features
```

```python
# Dataframe based on the scaled features
```

```python
cancer.head()
```

```python
cancer.columns
```

```python
cancer.columns[1:]
```

```python
df_feat = pd.DataFrame(Scaled_features,columns=cancer.columns[1:])
```

```python
df_feat
```

```python
X = df_feat
y = cancer['diagnosis']
```

```python
y
```

```python
#Split the data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25)
```

```python

# Build the KNN  Classifier  -- Model

from sklearn.neighbors import KNeighborsClassifier

# Initialize the Model

cancer_predit = KNeighborsClassifier(n_neighbors=40)     # k =
```

```python
# Train / fit the model
cancer_predit.fit(X_train,y_train)
```

```python
# Score
print (cancer_predit.score(X_test,y_test))
```

```python
prediction = cancer_predit.predict(X_test)
```

```python
# classification report & conf. matrix

from sklearn.metrics import classification_report
print (classification_report(y_test,prediction))
```

```python
#  Elbow method 
error_rate = []
for i in range(1,40):
    cancer_predit2 = KNeighborsClassifier(n_neighbors=i)
    cancer_predit2.fit(X_train,y_train)
    pred = cancer_predit2.predict(X_test)
    error_rate.append(np.mean(pred != y_test))
```

```python
error_rate
```

```python
# plot

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',
                marker='o',markerfacecolor='red',markersize=10)
plt.title('Error rate vs K value')
plt.xlabel('K')
plt.ylabel('Error rate')

plt.show()
```
